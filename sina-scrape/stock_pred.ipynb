```python
from bs4 import BeautifulSoup
import requests
from datetime import datetime, timedelta
from openpyxl import Workbook
import re

def stamp_divison(stamp):
    morning_start = datetime.strptime("09:30","%H:%M")
    morning_end = datetime.strptime("11:30", "%H:%M")
    afternoon_end = datetime.strptime("15:00", "%H:%M")  

    if stamp == afternoon_end or stamp == morning_end or stamp <= morning_start:
        return True
    else: 
        return False


def min_extract(date, symbol):
    """
    data: last day's date
    symbol: stock's symbol
    fifteen: if add 15min slot; 1 add 15min slot, 0 otherwise
    """
    
    wb = Workbook()
    # Or create a new named sheet and set it to active using index
    ws = wb.create_sheet(title="OldData")
    wb.active = 1  # Default sheet was 0, this new sheet is 1

    # Direct cell modification
    ws['A1'] = "日期"
    ws['B1'] = "时间"
    ws['C1'] = "开盘价"
    ws['D1'] = "收盘价"
    ws['E1'] = "最高价"
    ws['F1'] = "最低价"
    ws['G1'] = "成交量"
    ws['H1'] = "成交额"
    ws['I1'] = "买入量"
    ws['J1'] = "买入金额"

    # day loop
    for i in range(1):
        # init
        num_page = 1
        stamp = datetime.strptime("15:00", "%H:%M")
        price_ls = []
        prev_stamp = " "

        keys = ['open', 'closing', 'max', 'min', 'volumne', 'turnover', 'buyq', 'buyp']
        flow = {key: 0 for key in keys}
        
        # page loop
        while True:
            # download data from single page of date
            link = 'http://market.finance.sina.com.cn/transHis.php?symbol=' + symbol + '&date='+ date + '&page=' + str(num_page) + '-Original%20Source'
            print(link)
            html = requests.get(link)
            soup = BeautifulSoup(html.content, "lxml")
            table = soup.find("table", attrs={"class":"datatbl"})
            
            # next date
            # no value in cur date
            if table == None:
                # soup check 1
                html = requests.get(link)
                soup = BeautifulSoup(html.content, "lxml")
                table_check = soup.find("table", attrs={"class":"datatbl"})
                if table_check != None:
                    table = table_check
                else:
                    # soup check 2
                    html = requests.get(link)
                    soup = BeautifulSoup(html.content, "lxml")
                    table_check2 = soup.find("table", attrs={"class":"datatbl"})
                    if table_check2 != None:
                        table = table_check
                    else:
                        date = datetime.strptime(date, "%Y-%m-%d")
                        modified_date = date - timedelta(days=1)
                        date = datetime.strftime(modified_date, "%Y-%m-%d")
                        break
            # no value in cur page
            elif len(table.find_all("tr")[1:]) == 0:
                # no data available
                test = datetime.strptime("15:00", "%H:%M")
                if stamp == test:
                    pass
                else:
                    # update < 9:30 data
                    stamp_str = "09:30"
                    ws.append([date, stamp_str, price_ls[-1], price_ls[0], max(price_ls), min(price_ls), 
                            flow['volumne'], flow['turnover'], flow['buyq'], flow['buyp']])

                date = datetime.strptime(date, "%Y-%m-%d")
                modified_date = date - timedelta(days=1)
                date = datetime.strftime(modified_date, "%Y-%m-%d")
                break
            
            # extract data of current stamp
            for row in table.find_all("tr")[1:]:
                th_data = [th.get_text() for th in row.find_all("th")]                    
                td_data = [td.get_text() for td in row.find_all("td")]
                cur_stamp = datetime.strptime(th_data[0][:5], "%H:%M")
                cur_stamp_sec = th_data[0]

                if stamp_divison(stamp):
                    # purpose: skip next if statement
                    stamp = cur_stamp
                
                # check repeated rows
                if prev_stamp == cur_stamp_sec:
                    continue
                prev_stamp = cur_stamp_sec
                
                # different stamp
                if cur_stamp < stamp:
                    stamp_str = datetime.strftime(stamp, "%H:%M")
                    ws.append([date, stamp_str, price_ls[-1], price_ls[0], max(price_ls), min(price_ls), 
                        flow['volumne'], flow['turnover'], flow['buyq'], flow['buyp']])
                        
                    # update stamps
                    stamp = cur_stamp
                    flow = {key: 0 for key in keys}
                    price_ls = []   
                
                # price list
                if td_data[0] != None:
                    price_ls.append(float(td_data[0].replace(',','')))
                else:
                    print('ERROR: price')

                # update volumne and turnover
                if td_data[2] != None:
                    flow['volumne'] += float(td_data[2].replace(',',''))
                else:
                    print('ERROR: volumne')

                if td_data[3] != None:
                    flow['turnover'] += float(td_data[3].replace(',',''))
                else:
                    print('ERROR: turnover')

                # 买入量 买入总额 update
                if th_data[-1] == '买盘':
                    if td_data[2] != None:
                        flow['buyq'] += float(td_data[2].replace(',',''))
                    else: 
                        print('ERROR: buyq')

                    if td_data[3] != None:
                        flow['buyp'] += float(td_data[3].replace(',',''))
                    else:
                        print('ERROR: buyp')

                elif th_data[-1] == '中性盘':
                    if td_data[2] != None:
                        flow['buyq'] += float(td_data[2].replace(',',''))
                    else: 
                        print('ERROR: buyq')

                    if td_data[3] != None:
                        flow['buyp'] += (float(td_data[3].replace(',',''))/2)
                    else:
                        print('ERROR: buyp')

                elif th_data[-1] == '卖盘':
                    pass
                else: 
                    print('ERROR buysell')           

            num_page += 1
    
    file_name = symbol + '.xlsx'
    wb.save(file_name)  # Write to disk
    print("1分钟工作表下载完毕")
    
```

```python
from openpyxl import load_workbook

def fifteen_extract(filename):
    wb = load_workbook(filename)
    ws = wb["OldData"]
    ws15 = wb.create_sheet(title="15min")
    ws15['A1'] = "日期"
    ws15['B1'] = "时间"
    ws15['C1'] = "开盘价"
    ws15['D1'] = "收盘价"
    ws15['E1'] = "最高价"
    ws15['F1'] = "最低价"
    ws15['G1'] = "成交量"
    ws15['H1'] = "成交额"
    ws15['I1'] = "买入量"
    ws15['J1'] = "买入金额"

    keys = ['volumne', 'turnover', 'buyq', 'buyp']
    flow = {key: 0 for key in keys}
    price_max = []
    price_min = []
    open_price = ws['C2'].value
        
    stamp = datetime.strptime("15:00", "%H:%M")
    next_stamp = stamp - timedelta(minutes=15)
        
    for i in range(2, ws.max_row):
        time_cell = 'B' + str(i)
        cur_stamp = datetime.strptime(ws[time_cell].value, "%H:%M")
        
        while cur_stamp < next_stamp:
            if len(price_max) == 0:
                max_val = 0
                min_val = 0
                open_price = 0
                end_price = 0
            else:
                max_val = max(price_max)
                min_val = min(price_min)
            
            stamp_str = datetime.strftime(stamp, "%H:%M")
            ws15.append([ws['A'+str(i)].value, stamp_str, open_price, end_price, max_val, min_val, 
                flow['volumne'], flow['turnover'], flow['buyq'], flow['buyp']])
                
            stamp = next_stamp
            next_stamp = stamp - timedelta(minutes=15)
                    
            flow = {key: 0 for key in keys}
            price_max = []
            price_min = []
            open_price = ws['C'+str(i)].value
            
        # update 
        price_max.append(ws['E'+str(i)].value)
        price_min.append(ws['F'+str(i)].value)
        flow['volumne'] += ws['G'+str(i)].value
        flow['turnover'] += ws['H'+str(i)].value
        flow['buyq'] += ws['I'+str(i)].value
        flow['buyp'] += ws['J'+str(i)].value
        end_price = ws['D'+str(i)].value
    
    # last update
    stamp_str = datetime.strftime(stamp, "%H:%M")
    ws15.append([ws['A'+str(i)].value, stamp_str, open_price, end_price, max_val, min_val, 
            flow['volumne'], flow['turnover'], flow['buyq'], flow['buyp']])
    
    wb.save(filename)  # Write to disk
    print("15分钟工作表下载完毕")
```
